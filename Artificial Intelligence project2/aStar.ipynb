```python
from heapq import heappush, heappop
import copy
import tracemalloc
import time
```

```python
#Initialization
representMovesString = ["up" , "right" , "down" , "left"]
goal = [[1,2,3],[4,5,6],[7,8,0]]
samples = []
easy_samples = []
medium_samples = []
hard_samples = []
count = 0
totalTimeForAStartSearch = 0
totalMemoryForAStartSearch = 0
totalNumberOfAStartSearch = 0

class Status:
    def __init__(self, pState, maze, h, g, i, j, prevAction):
        self.pState = pState
        self.maze = maze
        self.h = h
        self.g = g
        self.i = i
        self.j = j
        self.action = prevAction
    def __lt__(self, next):
        return self.g + self.h < next.g + next.h
```

```python
#read Example.txt file and Initialization
with open("Examples.txt") as fp:
    while True:
        count += 1
        line = fp.readline()
  
        if not line:
            break
        samples.append(line.strip())
        
for i in range(len(samples)):
    if(i>=1 and i<21):
        easy_samples.append(samples[i])
    elif(i>=23 and i<38):
        medium_samples.append(samples[i])
    elif(i>=41 and i<61):
        hard_samples.append((samples[i]))

for i in range(len(easy_samples)): 
    easy_samples[i]=easy_samples[i][2:]
    easy_samples[i]=easy_samples[i][:17]
    easy_samples[i]=easy_samples[i].split(",")
    easy_samples[i]=[[eval(easy_samples[i][0]),eval(easy_samples[i][1]),eval(easy_samples[i][2])],[eval(easy_samples[i][3]),eval(easy_samples[i][4]),eval(easy_samples[i][5])],[eval(easy_samples[i][6]),eval(easy_samples[i][7]),eval(easy_samples[i][8])],]
for i in range(len(medium_samples)): 
    medium_samples[i]=medium_samples[i][2:]
    medium_samples[i]=medium_samples[i][:17]
    medium_samples[i]=medium_samples[i].split(",")
    medium_samples[i]=[[eval(medium_samples[i][0]),eval(medium_samples[i][1]),eval(medium_samples[i][2])],[eval(medium_samples[i][3]),eval(medium_samples[i][4]),eval(medium_samples[i][5])],[eval(medium_samples[i][6]),eval(medium_samples[i][7]),eval(medium_samples[i][8])],]
for i in range(len(hard_samples)): 
    hard_samples[i]=hard_samples[i][2:]
    hard_samples[i]=hard_samples[i][:17]
    hard_samples[i]=hard_samples[i].split(",")
    hard_samples[i]=[[eval(hard_samples[i][0]),eval(hard_samples[i][1]),eval(hard_samples[i][2])],[eval(hard_samples[i][3]),eval(hard_samples[i][4]),eval(hard_samples[i][5])],[eval(hard_samples[i][6]),eval(hard_samples[i][7]),eval(hard_samples[i][8])],]
```

```python
#methods

def find_zero(maze):
    for i in range(3):
        for j in range(3):
            if maze[i][j]==0:
                return [i,j]
def displaySeq(state):
    if state == None:
        return
    else:
        displaySeq(state.pState)
        print(state.action,end="")
        print("(",state.h+state.g,")",end="-")

def dist(prevState, ii, jj, n, action):
    updatedMaze = copy.deepcopy(prevState.maze)
    updatedMaze[prevState.i][prevState.j] = prevState.maze[ii][jj]
    updatedMaze[ii][jj] = prevState.maze[prevState.i][prevState.j]
    
    cost = 0
    for i in range(n):
        for j in range(n):
            for k in range(n):
                for p in range(n):
                    if (goal[k][p] == updatedMaze[i][j]):
                        g=k
                        h=p
            cost += abs(h - j) + abs(g - i)
    
    return Status(prevState, updatedMaze, cost , prevState.g + 1, ii, jj, action)
```

```python
# if f.h is equal to 0 then we solved the maze
# else iterate on actions
# move the zero or empty sec
# checking if that action is possible : # 
def aStar(startState):
    zeroPlace = find_zero(startState)
    i , j = zeroPlace[0] , zeroPlace[1]
    heapQueue = []
    
    
    cost = 0
    for ii in range(3):
        for jj in range(3):
            for k in range(3):
                for p in range(3):
                    if (goal[k][p] == startState[i][jj]):
                        g=k
                        h=p
            cost += abs(h - jj) + abs(g - ii)
    
    heappush(heapQueue, Status(None, startState, cost, 0, i, j, "first F score"))
    while True:
        if (len(heapQueue)==0):
            return
        f = heappop(heapQueue)
        if f.h == 0:
            return f
        for i in representMovesString:
            x, y = 0,0
            if (i == "right" or i == 'r'):
                x,y = f.i, f.j + 1
            elif (i == "down" or i == 'd'):
                x,y = f.i + 1, f.j
            elif (i == "left" or i == 'l'):
                x,y = f.i, f.j - 1
            elif (i == "up" or i == 'u'):
                x,y = f.i - 1, f.j
            
            if (x >= 0 and x < 3 and y >= 0 and y < 3):
                newState = dist(f, x, y, 3, i)
                heappush(heapQueue, newState)
```

```python
#astar for easy samples
print("for easy samples:  ")
for i in range(len(easy_samples)):
    displayMat = easy_samples[i]
    print("for puzzle: ",displayMat)
    
    #bottom two lines calculate the whole time and memory
    startAlgorithm = time.time()
    tracemalloc.start()
    
    f = aStar(easy_samples[i])
    
    totalTime = time.time() - startAlgorithm
    totalMemory = tracemalloc.get_traced_memory()[1]
    totalNumberOfAStartSearch += totalTime
    totalMemoryForAStartSearch += totalMemory
    totalNumberOfAStartSearch += 1
    print("time : ", totalTime)
    print("memory : ", totalMemory)
    
    print("action sequence: ",end="")
    displaySeq(f)
    print()
    print()

print()
print()
```
